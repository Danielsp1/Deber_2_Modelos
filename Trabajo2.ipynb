```python
df = df.rename(columns={'Label': 'Malicioso?'})
print(df)
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

X = df.iloc[:, 1:532]  
y = df['Malicioso?']  

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)


plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', s=50, alpha=0.5)
plt.title('PCA')
plt.xlabel('Componente principal 1')
plt.ylabel('Componente principal 2')
plt.colorbar(label='Malicioso?')
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.3, random_state=42)


print("Conjunto de entrenamiento:", X_train.shape, y_train.shape)
print("Conjunto de validación:", X_val.shape, y_val.shape)
print("Conjunto de prueba:", X_test.shape, y_test.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report


label_encoder = LabelEncoder()
df['Malicioso?'] = label_encoder.fit_transform(df['Malicioso?'])


X = df.drop('Malicioso?', axis=1)
y = df['Malicioso?']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


svm_model = SVC(kernel='linear', class_weight='balanced')  
svm_model.fit(X_train, y_train)


y_pred = svm_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Exactitud del modelo SVM:", accuracy)
print("\nReporte de clasificación:")
print(classification_report(y_test, y_pred))
```

```python

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import LabelEncoder
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report



label_encoder = LabelEncoder()
df['Malicioso?'] = label_encoder.fit_transform(df['Malicioso?'])


X = df.drop('Malicioso?', axis=1)
y = df['Malicioso?']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


param_grid = {
    'C': [0.1, 1, 10, 100],  # Parámetro de regularización
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],  # Tipo de kernel
    'gamma': ['scale', 'auto']  # Coeficiente del kernel (solo relevante para kernels rbf, poly y sigmoid)
}


svm_model = SVC(class_weight='balanced')


grid_search = GridSearchCV(svm_model, param_grid, cv=5, scoring='accuracy')


grid_search.fit(X_train, y_train)


best_params = grid_search.best_params_
print("Mejores hiperparámetros:", best_params)


y_pred = grid_search.predict(X_test)


accuracy = accuracy_score(y_test, y_pred)
print("Exactitud del modelo SVM optimizado:", accuracy)
print("\nReporte de clasificación:")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score, roc_curve
import matplotlib.pyplot as plt


y_val_pred = grid_search.predict(X_val)


accuracy_val = accuracy_score(y_val, y_val_pred)
precision_val = precision_score(y_val, y_val_pred)
recall_val = recall_score(y_val, y_val_pred)
roc_auc_val = roc_auc_score(y_val, y_val_pred)

print("Métricas de evaluación en el conjunto de validación:")
print("Exactitud:", accuracy_val)
print("Precisión:", precision_val)
print("Sensibilidad (Recall):", recall_val)
print("AUC-ROC:", roc_auc_val)


fpr_val, tpr_val, _ = roc_curve(y_val, y_val_pred)


y_test_pred = grid_search.predict(X_test)


accuracy_test = accuracy_score(y_test, y_test_pred)
precision_test = precision_score(y_test, y_test_pred)
recall_test = recall_score(y_test, y_test_pred)
roc_auc_test = roc_auc_score(y_test, y_test_pred)

print("\nMétricas de evaluación en el conjunto de prueba:")
print("Exactitud:", accuracy_test)
print("Precisión:", precision_test)
print("Sensibilidad (Recall):", recall_test)
print("AUC-ROC:", roc_auc_test)


fpr_test, tpr_test, _ = roc_curve(y_test, y_test_pred)


plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(fpr_val, tpr_val, color='blue', label='ROC curve (area = %0.2f)' % roc_auc_val)
plt.plot([0, 1], [0, 1], color='red', linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve (Validation)')
plt.legend(loc="lower right")

plt.subplot(1, 2, 2)
plt.plot(fpr_test, tpr_test, color='green', label='ROC curve (area = %0.2f)' % roc_auc_test)
plt.plot([0, 1], [0, 1], color='red', linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve (Test)')
plt.legend(loc="lower right")

plt.tight_layout()
plt.show()
```

```python

coeficientes = grid_search.best_estimator_.coef_[0]


coef_df = pd.DataFrame({'Feature': X.columns, 'Coefficient': coeficientes})


coef_df['Abs_Coefficient'] = abs(coef_df['Coefficient'])
coef_df = coef_df.sort_values(by='Abs_Coefficient', ascending=False)

print("Las características más importantes para la clasificación de malware son:")
print(coef_df.head(10))
```

```python
top_features = coef_df.head(10)


coef_percentages = top_features['Coefficient'] * 100


colors = ['red' if coef < 0 else 'green' for coef in top_features['Coefficient']]


plt.figure(figsize=(10, 6))
bars = plt.barh(top_features['Feature'], coef_percentages, color=colors)
plt.xlabel('Coeficiente (%)')
plt.ylabel('Característica')
plt.title('Coeficientes de las características más importantes')
plt.gca().invert_yaxis()  


for bar, coef in zip(bars, top_features['Coefficient']):
    plt.text(bar.get_width() * (1 if coef > 0 else -1),
             bar.get_y() + bar.get_height() / 2,
             f'{coef:.2f}%',  
             va='center', ha='center',
             color='white' if abs(coef) > 10 else 'black')  


plt.show()

```
